```python
import csv
import emoji

# Baca data dari 'komentar.csv'
with open('komentar.csv', mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)
    komentar = [row[0] for row in reader]  # Asumsikan kolom pertama berisi teks

# Siapkan untuk menyimpan komentar yang telah diganti emoji-nya
with open('komentar_output.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Teks'])  # Menulis header CSV

    # Loop melalui setiap komentar untuk mengganti emoji
    for text in komentar:
        # Mengganti emoji dengan deskripsi teks
        text_with_descriptions = emoji.demojize(text, language='id')  # Gunakan bahasa Indonesia

        # Simpan teks yang sudah diubah ke dalam file baru
        writer.writerow([text_with_descriptions])  # Simpan teks komentar ke dalam file CSV
        print(text_with_descriptions)  # Cetak teks komentar ke konsol
```
